```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt 
```

```python
os.chdir(r"c:\Users\admin\Desktop\research")    # use os library to set parent directory

file_extension = ".csv"
all_file_name = [i for i in glob.glob(f"*\*{file_extension}")]  # use glob library to get all file names
```

original data frame:
    index   |   ID  |   value   |   timestamp

- sample rate: approximately per second

new data frame:
    date |   values of ID (4 kinds in total = 4 columns)

- index: date
- sample rate: per day
- value: mean of that day

```python
list_of_df = []

for file_name in all_file_name[0:5]:    # just test with first 5 file

    df = pd.read_csv(file_name, sep = '\t', engine = 'python', header = None, names = ["ID", "value", "timestamp"]) # original data frame (file)

    df = df.pivot_table(values = 'value', columns = 'ID')   # default of aggfunc is mean()

    df['date'] = [file_name[5:15]]  # add date column by getting it from file name

    list_of_df.append(df)

df = pd.concat(list_of_df).set_index('date')    

print(df)
```
